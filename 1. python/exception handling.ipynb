# Custom Exception handling

```python
age = int(input("enter your age: "))
# user can enter a negative value and the system will not give an error
# but we know that age can never be negative
```

```python
class ValidateAge(Exception):

    def __init__(self,msg):
        self.age=msg
```

```python
def validateage(age):
    if age<0:
        raise ValidateAge("Age cannot be negative")
    
    elif age>100:
        raise ValidateAge("Age cannot be greater than 100")
    else:
        print("Age is valid")
```

```python
try:
    age = int(input("enter your age: "))
    validateage(age)
except ValidateAge as e:
    print(e)
```

# Best Practice Exception handling

```python
# use always a specific exception
# dont use as shown below
try:
    10/0
except Exception as e:
    print(e)
```

```python
try:
    10/0
except ZeroDivisionError as e:
    print(e)
```

```python
# print always a proper message
try:
    10/0
except ZeroDivisionError as e:
    print("this is my zero division error i am handling",e)
```

```python
# always try to log
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    10/0
except ZeroDivisionError as e:
    logging.error("this is my zero division error i am handling {}".format(e))
```

```python
# always avoid to write a multiple exception handling
try:
    10/0
except FileNotFoundError as e:
    logging.error("this is my file not found error i am handling {}".format(e))
except AttributeError as e:
    logging.error("this is my attribute error i am handling {}".format(e))
except ZeroDivisionError as e:
    logging.error("this is my zero division error i am handling {}".format(e))
```
