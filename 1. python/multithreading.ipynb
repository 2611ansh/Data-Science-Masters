```python
import threading
```

```python
def test(id):
    print("this is my test id %d"%id)
```

```python
test(10)
```

```python
test(1)
```

```python
test(3)
```

```python
thred=[threading.Thread(target=test,args=(i,)) for i in [10,1,3]]
#args=(i,) is providing imput i.e. i can provide as many inputs i want 
```

```python
for t in thred:
    t.start()
```

```python
shared_var=0
lock_var=threading.Lock()
def test3(x):
    global shared_var
    with lock_var:
        shared_var=shared_var + 1
        print("value of x %d and value of shared_var %d" %(x,shared_var))
        #time.sleep(1)
    
thread5=[threading.Thread(target=test3,args=(i,)) for i in [1,2,3,4,5]]
for t in thread5:
    t.start()
```
