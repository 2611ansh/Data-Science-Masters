```python
class bank_account:
    
    def __init__(self,balance):
        self.__balance =balance
    
    def deposit(self,amount):
        self.__balance=self.__balance + amount
    
    def withdraw(self,amount):
        if self.__balance >= amount:
            self.__balance=self.__balance - amount
            return True
        else:
            return False
    
    def get_balance(self):
        return self.__balance
```

```python
obj_bank_account = bank_account(1000)
```

```python
obj_bank_account.get_balance()
```

```python
obj_bank_account.deposit(4000)
```

# decorator

```python
def test():
    print("start")
    print(1+1)
    print('end')
```

```python
test()
```

```python
def deco(func):
    def inner_deco():
        print("start")
        func()
        print("end")
    return inner_deco
```

```python
def test():
    print(1+1)
```

```python
test()
```

```python
@deco #to return the same start and beginning everytime
def test():
    print(1+1)
```

```python
test()
```

```python
import time
def timer_test(func):
    def timer_test_inner():
        start = time.time()
        func()
        end=time.time()
        print(end-start)
    return timer_test_inner
```

```python
@timer_test
def test():
    print(1+1)
```

```python
test()
```

```python
@timer_test
def test():
    for i in range(1000000):
        pass
```

```python
test()
```

```python
class pwskills:
    def __init__(self,course_price,course_name):
        self.__course_price=course_price
        self.course_name=course_name

    @property #accesser
    def course_price_access(self):
        return self.__course_price

#external user using below function will be able to set price which is private
    @course_price_access.setter #setter 
    def course_price_set(self,price):
        if price <=3500:
            pass
        else:
            self.__course_price=price

#external user able to delete private and protected variable    
    @course_price_access.deleter
    def course_price_del(self):
        del self.__course_price
```

```python
pw=pwskills(3500,"DSM")
```

```python
pw.course_price_access
```

```python
del pw.course_price_del
```
