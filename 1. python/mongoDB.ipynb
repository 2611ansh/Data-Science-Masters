```python
#to see all the records inside my collection
for i in coll_create.find():
    print(i)
```

```python
#to see the first record inside my collection
coll_create.find_one()
```

```python
#to find record on basis of some condition
for i in coll_create.find({'full_name' : 'jaspreet'}):
    print(i)
```

```python
# $gt : greater then
# $get : greater then or equal to
for i in coll_create.find({'phone_no' : {'$gte' : 9400000000}}):
    print(i)
```

```python
#update data
#update_one will update the first record matching the condition
#update_many will update all the records
# $set : set new record
coll_create.update_many({'name' : 'ansh'},{'$set':{'name' : 'jaspreet'}})
```
