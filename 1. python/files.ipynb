# buffer read and write

```python
import io
```

```python
with open("test.txt",'wb') as f:
    file = io.BufferedWriter(f)
    file.write(b"this is my buffer write\n")
    file.write(b"this is my second line that is am trying to write ")
    file.flush() #to close a buffer write file
```

```python
with open("test.txt","rb") as f:
    file = io.BufferedReader(f)
    data=file.read()
    print(data)
```

```python
with open("test.txt","rb") as f:
    file = io.BufferedReader(f)
    data=file.read(20)
    print(data)
```

```python
l=[1,2,3,4,[4,5,6],"ansh","jas"]
```

```python
l1_int = []
l2_str = []
for i in l:
    logging.info("this is the start of my first for loop {}".format(l))
    logging.info("this is the value of i am logging {}".format(i))
    if type(i) ==list:
        for j in i:
            logging.info("logging my j{j} and i is {i}".format(i=i,j=j))
            if type(j) == int:
                l1_int.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else:
        if type(i)==str:
            l2_str.append(i)
logging.info("this is my final result withh all int {l2} , with all str{l2}".format(l1=l1_int,l2=l2_str))
```
